```python
import numpy as np
import matplotlib.pyplot as plt

x_min = y_min = -3
x_max = y_max = 3
step = 0.01

x, y = np.meshgrid(np.arange(x_min, x_max, step), np.arange(y_min, y_max, step))

eq1 = np.sin(x+2) - y - 1.5
eq2 = x + np.cos(y - 2) - 0.5

fig, ax = plt.subplots(figsize=(10, 10))

ax.contour(x, y, eq1, levels=[0], colors="red")
ax.contour(x, y, eq2, levels=[0], colors="blue")

ax.set_xlim([x_min, x_max])
ax.set_ylim([y_min, y_max])
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.grid(True)
plt.show()
```

Метод простої ітерації

```python
eq1 = lambda x : np.sin(x+2) - 1.5
eq2 = lambda y: (np.cos(y - 2)*-1) + 0.5

x0 = 1
y0 = -2
e = 0.0001

xn = x0
yn = y0
xn1 = eq2(yn)
yn1 = eq1(xn)

while abs(xn1 - xn) >= e and abs(yn1 - yn) >= e:
  xn = xn1
  yn = yn1
  xn1 = eq2(yn)
  yn1 = eq1(xn)

print("x =", xn1)
print("y =", yn1)

```
